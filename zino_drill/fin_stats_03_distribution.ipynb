```python
def bin_disk(k, n, p) : 
    nck = factorial(n)/ (factorial(k) * factorial(n-k)) # Combination
    proba_df = nck * pow(p,k) * pow(1-p, n-k) # p의확률 k 번, 1-p 의 확률 n-k번 
    return proba_df
```

```python
def bin_dist(k, n, p) : 
    from math import factorial 

    nck = factorial(n)/ (factorial(k) * factorial(n-k))
    proba_df = nck * pow(p, k) * pow(1-p, n-k)
    return proba_df 


def binomial_df(k, n, p) : 
    import numpy as np
    import matplotlib.pyplot as plt

    x = np.arange(k)
    proba_df1 = np.array([bin_dist(k,n,p) for k in range(k)])  # QQQQQQQQQQQQQQQQQQQQQQQQQ -> np.array([~]) 무슨 의미지, 반복문인가..? 
                                                            # 반복문 맞음 
                                                                # k 0 ~ k-1로 k개의 각 성공확률의 확률 분포인가  맞더라도 왜 이런형태인지는 모름
                                                                
    print(proba_df1)


    plt.ylim(0, p)
    plt.xlim(-1, k)
    plt.bar(x, proba_df1, label = f'independent experiments : {n} \n Success proba for n = 1 : \
    {p*100}%')
    plt.xlabel('independent experiments')
    plt.ylabel('proba of success') 
    plt.legend()
    plt.show()   

    
```

```python
#무수히 많이 시행한 경우, 실제 베르누이 분포 시각화 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import pandas as pd
import numpy as np

p = 0.3
n = 1000
rv = bernoulli(p)
X = rv.rvs(n, random_state = 2010) # rvs : 무작위 샘플 생성 함수 , random_state  = 무작위 샘플 생성할때 고유한 시드 
# print(X)

y = np.bincount(X, minlength=2)/len(X)
XX = [0,1]

sns.countplot(x = X)
plt.xticks([0,1], [f"X=0\n{int(y[0]*n)} times", f"X=1\n{int(y[1]*n)} times"])
plt.show()

df = pd.DataFrame({'theoretic' : rv.pmf(XX), 'simulation': y}).stack()
df = df.reset_index()
df.columns = ['value', 'type', 'ratio']
df.pivot('value', 'type', 'ratio')
print(df)

sns.barplot(x='value', y='ratio', hue= 'type', data = df)
plt.show()
```

```python
#함수화

def sim_bernoulli_df(p,n) :
    import pandas as pd
    import numpy as np
    import seaborn as sns
    from scipy.stats import bernoulli

    rv = bernoulli(p)
    X = rv.rvs(n, random_state = 2022)
    y = np.bincount(X, minlength=2)/len(X)

    sns.countplot(x=X)
    plt.xticks([0,1], [f"X=0\n{int(y[0]*n)} times", f"X=1\n{int(y[1]*n)} times"])
    plt.title(f"simulation of Bernoullidf ({n} times)")
    plt.show()

    df = pd.DataFrame({'theoretic' : rv.pmf(XX), 'simulation' : y}).stack()
    df = df.reset_index()
    df.columns = ['value', 'type', 'ratio']
    df.pivot('value', 'type', 'ratio')
    print(df)

    sns.barplot(x='value', y='ratio', hue='type', data = df)  # hue가 다른 bar그래프 색깔을 달리해서 붙여서 보여줌
    plt.title(f"simulation vs theoretic({n} times)")
    plt.show()

    
sim_bernoulli_df(0.4, 10000)
```

## A. Normal Distribution & Standard Normal Distribution


```python
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 정규성 확보 

N1 = 500
sigma_1 = 10
mu_1 = 10

# random data

np.random.seed(0)
x1 = sp.stats.norm(mu_1, sigma_1).rvs(N1)
```

```python
# visualizing

# kde = Kernel Density Estimation : 데이터의분포를 부드러운 곡선으로 continuously
    # True로 바꿀시, continuous로 fit = stats.norm과 비교가능 
# fit= stats.norm : 정규분포 곡선을 표시하여, 데이터의 분포가 정규분포를 따르는지 확인 가능
# np.round ~ : x1의 standard deviation을 소수점 셋째자리까지 반올림
ax = sns.distplot(x1, kde = True, fit= sp.stats.norm, label = f"normal df(Variance : {np.round(x1.std(), 3)})")
plt.legend(bbox_to_anchor = (1.0, .2)) # chart 가로 세로 크기 
plt.show()
```

```python
# 정규분포 & 표준정규분포


def normal_df(N, sigma, mu) : 
    plt.rcParams['axes.unicode_minus'] = False # 음수 글자 깨질때 하는 설정 
    plt.rc('font', family = 'Malgun Gothic')
    # 랜덤데이터 - Normal distribution

    np.random.seed(0)
    x = sp.stats.norm(mu, sigma).rvs(N) # rvs = random variates of given type
    
    sns.displot(x-1, kde = False, label = f"normal df(Variance) : {x.std() : .3f}")
    plt.legend(bbox_to_anchor = (1.0, .2))
    plt.ylabel('PDF(확률밀도함수)')
    plt.xlabel('x(독립변수)')
    if (mu ==1) & (sigma == 1) :
        plt.title('표준정규분포')
    else : 
        plt.title('정규분포')
    plt.show()

normal_df(1000000, 10,10)

normal_df(1000000, 1, 1)
```

## B. T분포

```python
import math
```

```python
# t 정의 : (-5,5까지 101개의 값으로 분류 -> 0.1 간격)
t = np.linspace(-5,5,500)

# degree of freedom

DF = 10
# math.gamma함수를 이용하여 t분포 정의
y = (math.gamma((DF+1)/2)/(math.gamma(DF/2)*math.sqrt(math.pi * DF))) * (1+t**2/DF)**(-(DF+1)/2)

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('pdf')
plt.title('t-df')
plt.legend([f"t-df (DF = {DF})"])
plt.show()
```

```python
# Scipy로도 t분포 그릴 수 있어 

DF = 10 
t = np.linspace(-5,5, 501)
y = stats.t(DF).pdf(t)

plt.plot(t,y)
plt.xlabel('t')
plt.ylabel('pdf')
plt.title("t-df w/ scipy")
plt.legend(f't-df(DF = {DF})')
plt.show()
```

```python
def t_df_stats(DF):
    import scipy.stats as stats
    import numpy as np
    import matplotlib.pyplot as plt

    t = np.linspace(-5, 5, 101)
    y1 = stats.t(DF).pdf(t)
        
    plt.plot(t, y1)
    plt.xlabel("t")
    plt.ylabel("pdf")
    plt.title("t-df with scipy")
    plt.legend([f"t-df (Degree of Freedom = {DF})"])
    plt.show()
    
t_df_stats(10)


# QQQQQQQQQQQQQQQQQQQQQQQ 
    # 자유도 다른 pdf 동시에 그릴 수는 없나 ? 자유도에 따른 모양 비교 하고 싶음 
```

## C. F distribution



```python
import scipy.special as sp
import math
import numpy as np

# X축 값 설정 (0,5 사이에 일정한 거리로 1000개 데이터 입력)
x = np.linspace(0, 5, 1000)
# 각각의 독립적인 확률변수 데이터가 자유도 4, 10인 카이제곱 분포를 따르는 경우를 가정
nu1 = 4
nu2 = 10

# F분포를 따르는 확률변수를 생성
y = []
for f in x:
    y.append(( np.sqrt( ((nu1 * f) ** nu1 * nu2 ** nu2) / ((nu1 * f + nu2) ** (nu1 + nu2)) )) / 
            (f * sp.beta(nu1 / 2, nu2 / 2)))
    
plt.plot(x, y, label=f'F({nu1}, {nu2})')                       
plt.xlabel('F')
plt.ylabel("y")
plt.title('F 분포 without scipy')
plt.legend()
plt.show()
```

```python
import scipy.stats as stats

# 카이제곱분포를 따르는 확률변수의 자유도가 nu1, nu2일 때 F분포의 확률밀도함수는 y1으로 정의된다.
y1 = stats.f(nu1, nu2).pdf(x)        


plt.plot(x, y1, color="g", label='F(4, 50)')             
plt.xlabel('F')
plt.ylabel("y")
plt.title('scipy를 이용한 F 분포')
plt.legend()
plt.show()
```

```python
y1 = stats.f(nu1, nu2).pdf(x)        

plt.figure(figsize=(16,10))
plt.plot(x, y, alpha=.5, color="k", label=f'F({nu1}, {nu2}) without scipy')  
plt.plot(x, y1, alpha=.5, label=f'F({nu1}, {nu2}) with scipy')                                  
plt.xlabel('F')
plt.ylabel("y")
plt.title('F 분포')
plt.legend()
plt.show()
```

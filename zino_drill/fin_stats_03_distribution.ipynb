```python
import pandas as pd
```

```python
def bin_disk(k, n, p) : 
    nck = factorial(n)/ (factorial(k) * factorial(n-k)) # Combination
    proba_df = nck * pow(p,k) * pow(1-p, n-k) # p의확률 k 번, 1-p 의 확률 n-k번 
    return proba_df
```

```python
n =15
p = 0.3

x = np.arange(16)
print(x)
```

```python
proba_df1 = np.array([bin_disk(k, n, p) for k in range(16)])
print(proba_df1)
```

```python
def bin_dist(k, n, p) : 
    from math import factorial 

    nck = factorial(n)/ (factorial(k) * factorial(n-k))
    proba_df = nck * pow(p, k) * pow(1-p, n-k)
    return proba_df 


def binomial_df(k, n, p) : 
    import numpy as np
    import matplotlib.pyplot as plt

    x = np.arange(k)
    proba_df1 = np.array([bin_dist(k,n,p) for k in range(k)])  # QQQQQQQQQQQQQQQQQQQQQQQQQ -> np.array([~]) 무슨 의미지, 반복문인가..? 
                                                            # 반복문 맞음 
                                                                # k 0 ~ k-1로 k개의 각 성공확률의 확률 분포인가  맞더라도 왜 이런형태인지는 모름
                                                                
    print(proba_df1)


    plt.ylim(0, p)
    plt.xlim(-1, k)
    plt.bar(x, proba_df1, label = f'independent experiments : {n} \n Success proba for n = 1 : \
    {p*100}%')
    plt.xlabel('independent experiments')
    plt.ylabel('proba of success') 
    plt.legend()
    plt.show()   

    
```

```python
binomial_df(16, 15, 0.3)  # k=16, n = 15 가 가능한가? 15번 시행 중 16번 성공할 확률인데 QQQQQQQQQQQQQQQQQQQQQQQQQQ
                            # 계산은 (0.3)^16   *  (0.7)^(-1) 로 되었겠지만 
                            # 암튼 이건 왜 숫자 하나가 아니라 여러개가 나오는 거지 
```

```python
from scipy.stats import bernoulli
```

```python
# 동전던지기

# 확률설정

p = 0.3

rv = bernoulli(p)  # bernoulli(p) : p의 확률을 가진 베르누이 시행 

XX = [0,1]

plt.bar(XX, rv.pmf(XX), align='center')  # bernoulli(p).pmf(k) : p 확률을 가진 이항변수에서 k번 성공할 확률계산 
                                        # pmf = prob. mass func. 확률질량함수 (discrete variable)
                                        # QQQQQQQQQQQQQQQQQQQ -> k번 이라는데  pmf()안에는 list형태로 0,1 들어가면 인자가 순서대로 
plt.ylim(0,1)
plt.xticks([0,1], ['X=0', 'X=1'])
plt.ylabel('Proba')
plt.title('Bernoulli df')
plt.show()
```

```python
# 함수화 

def bernoulli_df(p) : 
    import matplotlib.pyplot as plt
    from scipy.stats import bernoulli

    proba = bernoulli(p)

    x = [0, 1]
    

    # 시각화

    plt.bar(x, proba.pmf(x), align = 'center')
    plt.ylim(0,1)
    plt.xticks([0,1], ['X=0(실패)', 'X=1(성공)'])
    plt.ylabel('Proba')
    plt.title("Bernoulli df")
    plt.show()

bernoulli_df(0.1)
```

```python
#무수히 많이 시행한 경우, 실제 베르누이 분포 시각화 
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import pandas as pd
import numpy as np

p = 0.3
n = 1000
rv = bernoulli(p)
X = rv.rvs(n, random_state = 2010) # rvs : 무작위 샘플 생성 함수 , random_state  = 무작위 샘플 생성할때 고유한 시드 
# print(X)

y = np.bincount(X, minlength=2)/len(X)
XX = [0,1]

sns.countplot(x = X)
plt.xticks([0,1], [f"X=0\n{int(y[0]*n)} times", f"X=1\n{int(y[1]*n)} times"])
plt.show()

df = pd.DataFrame({'theoretic' : rv.pmf(XX), 'simulation': y}).stack()
df = df.reset_index()
df.columns = ['value', 'type', 'ratio']
df.pivot('value', 'type', 'ratio')
print(df)

sns.barplot(x='value', y='ratio', hue= 'type', data = df)
plt.show()
```

```python
#함수화

def sim_bernoulli_df(p,n) :
    import pandas as pd
    import numpy as np
    import seaborn as sns
    from scipy.stats import bernoulli

    rv = bernoulli(p)
    X = rv.rvs(n, random_state = 2022)
    y = np.bincount(X, minlength=2)/len(X)

    sns.countplot(x=X)
    plt.xticks([0,1], [f"X=0\n{int(y[0]*n)} times", f"X=1\n{int(y[1]*n)} times"])
    plt.title(f"simulation of Bernoullidf ({n} times)")
    plt.show()

    df = pd.DataFrame({'theoretic' : rv.pmf(XX), 'simulation' : y}).stack()
    df = df.reset_index()
    df.columns = ['value', 'type', 'ratio']
    df.pivot('value', 'type', 'ratio')
    print(df)

    sns.barplot(x='value', y='ratio', hue='type', data = df)  # hue가 다른 bar그래프 색깔을 달리해서 붙여서 보여줌
    plt.title(f"simulation vs theoretic({n} times)")
    plt.show()

    
sim_bernoulli_df(0.4, 10000)
```

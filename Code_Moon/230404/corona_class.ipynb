```python
# 컬럼 이름 변경
# 컬럼 변수를 출력 

corona.columns  # 데이터 형태 리스트네 -> 리스트 형태로 이름 변경 
                # 컬럼 수 9개이기에, 변경 리스트 길이도 똑같이 맞춰줘야

```

```python
corona.columns = ['등록일시', '누적사망자', '누적확진자','게시글 번호', '기준일', '기준시간', '수정일시', '누적의심자', '누적확진률']  

# = 등호 까먹지 말고

```

```python
corona.head(2)  # 변경 완료띠
```

```python
# 컬럼 정렬 변경 
# index 또는 values 기준으로 바꿔보자 : sort_index    or     sort_values
# 보통 sort_index 에서 index는 1줄이기에 
# sort_values 는 ㅈㄴ 여러개자나 -> ('컬럼이름')들어가서 그 컬럼 기준으로 정렬 바꿀거얌

corona.sort_values('등록일시')
```

```python
# 데이터프레임에 정렬을 변경하는 함수
# sort_values()
# 매개변수 ascending : True (<-기본값) <- 오름차순 ,  False로 바꾸면 내림차순
# 매개변수 inplace : 기준이 되는 데이터프레임을 수정할 것인가를 지정 
                    # inplace의 기본값은 False : 기준 데이터프레인 수정 안해 -> True : 수정할거얏

corona.sort_values('등록일시', ascending = True, inplace = True)  # 안나오네? corona 자체 즉 기준변수의 내용을 진짜 바꿀꺼야 ! 
                                                                # 에러가 나지않고 시행이 잘되었어? -> ㅇㅋ 수정 잘되어버렷 오름차순으로 !

                                                            
```

```python
corona.head(10)  # 오름차순 잘 되었네? 과거의 데이터부터 순서대로 !! 
```

```python
corona.sort_index() # 다시 index 기준 오름차순으로 기기 
```

```python
# 인덱스 초기화 

# reset_index() -> 얘도 매개변수 2가지
# 매개변수 drop : 기본값(False)  -> 기존의 인덱스의 값을 삭제할지 지정 
# 매개변수 inplace : 기준이 되는 데이터프레임을 수정할지 지정 False

#corona.reset_index()  # 인덱스는 0~819 순으로 초기화 되었는데, 기준이 되는 819~0순 인덱스가 써있네? 

#corona.reset_index().drop('index', axis = 1) -> 밑에꺼가 기억이 안나면, 이렇게 써도 돼 똑같은거야. 
corona.reset_index(drop=True, inplace = True) # 짜잔 결과 안나오고 시행은 잘되네? 기준값 corona 변경 완료
```

```python
corona.tail(5)
```

```python
# 파생 변수

# 누적확진자 델타값으로 신규확진자 수 뽑아내볼까용? 

```

```python
# 데이터간의 차이를 출력하는 함수 : diff(n) , shift()

# diff(n)         : n칸 위로 가서 그 값과의 차이 
# n 기본값은 1 -> 하나 전 인덱스 데이터를 선택
# if n =2 -> 두개 전 인덱스 데이터를 선택 

corona.누적확진자 
# corona['누적확진자']
```

```python
corona.누적확진자.diff()  # 인덱스 0 은 전 값이 없어서 NaN
```

```python
# 이제 이 값을 넣어줘야지

corona['일일확진자'] = corona.누적확진자.diff()
```

```python
corona.tail(5)
```

```python
# shift()
# 이동한 값을 출력

corona2 = corona.head(5)
```

```python
corona2['누적사망자']
```

```python
corona2['누적사망자'].shift(1)  #값이 한 칸씩 이동
```

```python
# 일일사망자 파생변수 생성
corona['누적사망자'] - corona['누적사망자'].shift(1)
```

```python
# 추가해줘야지? 
corona['일일사망자'] = corona['누적사망자'] - corona['누적사망자'].shift(1)
```

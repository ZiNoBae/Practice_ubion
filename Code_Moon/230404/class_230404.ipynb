# 결측치
---

```python
import pandas as pd
import numpy as np
```

```python
_value = [
    [1,2,3],
    [4,np.nan,6],   #numpy안에 있는 nan의 결측치를 의미
    [7,8,np.nan]
]

df = pd.DataFrame(_value)
df
```

```python
# 결측치 확인 방법
# 데이터프레임의 정보를 확인 .info

df.info()

# 아래 셀 설명 - total 3 columns인데, non-null인 데이터가 3,2,2개다. 즉, 0,1,1개 씩 결측치가 있다.의 의미
# info를 통해서, index & column 수 및 결측치 및 데이터 타입 까지 확인해보는게 좋다
```

```python
# is()함수를 사용하여 결측치를 확인
# isna() 라는 함수 사용 ! 
# 결측치인 경우 True로 나옴. 
# 그렇담 데이터 많을때, True 갯수 언제세고 앉아있엉

df.isna()
```

```python
# True = 1 , False = 0 을 의미하니, 데이터 값 sum()으로 더하면 그 수만큼 결측치 있다는 의미. 
# 각 index 별 결측치 갯수

df.isna().sum()
```

```python
# 결측치를 제거 (보통 5%미만 일때)
# 결측치 지울 때는 신중할 것 -> 확보 데이터 대비 많이 제거되어 좋은 분석이 안될 수도. 
# dropna() : 결측치를 제거하는 함수 
# 매개변수 axis : '0', 'rows' 대입시, 결측치가 있는 "행"을 삭제 ㅁ
#               & '1', 'columns' 대입시, 결측치가 있는 "열"을 삭제



df.dropna(axis = 1)
```

```python
# sales 데이터 로드

sales = pd.read_csv('../csv/Sales Records.csv')
sales.head(5)
```

```python
flg = sales['Region'] == 'Asia'   # sales라는 데이터 프레임 중 'Region' 컬럼 한줄 선택 : 데이터 형태 시리즈가 되었고 'Asia'이면 True, 다르면 False
```

```python
# 데이터 프레임 필터 기능 
# loc[행의 조건, 열의 조건]   --location

# 행의 조건 필터
#   열은 다보고 행에만 조건 걸고 싶을 때는, 열의조건 생략 가능
#sales.loc[0] 하면 전체나옴 
#sales.loc[flg]

sales.loc[~flg]   # ~ 은 True & False 반전 -> 'Asia' 가 아닌 값만 ! 

#  열의 조건 필터

sales.loc[:,['Order Priority', 'Order Date']]   # 데이터 형태가 '데이터프레임'

# 행과 열의 조건 Both 필터

sales.loc[flg, ['Region']]


# iloc[행위 위치, 열의 위치]    --index  location

sales.iloc[1:3, 1:3]   # index의 값과 상관없이 index의 위치에 따라 결과값이 다르게 나옴
                    # 만일 위 cell에서 index들 순서 바뀌는 경우 등



# sales.loc[1:3]   # loc는 이렇게 입력하면 index의 "값"기준으로 ! 
```

```python
sales[['Region']]  # 리스트 형태라면 원소가 0개부터 가능하기에 시리즈가 아니라 데이터 프레임 형태로 ..? --Question

```

```python
# 결측치 포함된 데이터만 필터
# 결측치가 포함된 data 전부를 보고싶어? -> 행의 조건만 설정하면 되겠지. 

df.columns = ['A', 'B', 'C']  # 컬럼이름 바꾸기

flg_na = df.isna().any(axis=1)  # 한개라도 True 있으면, True라 출력


```

```python
# 결측치만 일단 확인해보자고 통일성도 없고 개똥 데이터이지만

uriage = pd.read_csv("../csv/uriage.csv")

uriage.info()
```

```python
# 결측치 확인

uriage.isna().sum()
```

```python
# 결측치가 존재하는 조건 생성
uriage.isna().any(1)  #필터에 쓰려고 이거 뽑아놓은거야 이건 행의 조건 결과값 시바 2999개자나 행 보여주는 중이야 
                        # 행중에 하나라도 결측치 있으면 True
```

```python
uriage.loc[~(uriage.isna().any(1))]  # ~ 이렇게 하려면 ~() 이렇게 써서 하는걸 습관화 시킬 것. 지금은 나오는데 나중에 Error 나올 수도 있대
```

```python
uriage.dropna(axis = 0)  #  결측치가 있는 행을 삭제 dropna -> axis = 0 ... cf) any -> axis(1) 반대네 
```

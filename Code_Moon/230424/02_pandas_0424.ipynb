## 할로윈 투자전략 

1. 11월1일에 buy -> 6개월간 유지 -> 4월 말일에 sell
2. 11월 첫날 시가로 구매
3. 4월 마지막날에 종가로 판매 
4. 수익률 = (3번과정 - 2번과정)/ 2번과정
5. csv/AMZN.csv 데이터로 시행
    - 2000년부터 2019년까지의 데이터로 백테스팅
    

ZiNo 

1. From df 
    - 월 마지막날 dataframe : df_last -> 이후 4월만 뽑아서 : df_4 
    - 월 첫날 dataframe : df_first -> 이후 11월만 뽑아서 : df_11
2. df_final : df_4 & df_11 concat 후에 날짜순 재정렬
3. df_final 중 4월에만 종가와 전년 11월 시가 비교하여 return 컬럼 채움(나머지는 0)
4. return컬럼 중 0인애들 제외 -> df_rtn
5. return 컬럼에 1더한 total_rtn 컬럼 만든 후 cumprod()

```python
df_4 = pd.DataFrame()
```

```python
for i in range(0, len(df_last), 1) : 
    if df_last['STD-YM'][i].endswith('04') :
        df_4 = pd.concat([df_4, df_last.iloc[[i]]])
df_4
```

```python
# 11월 첫날 
df_first = df.loc[df['STD-YM'] != df.shift(1)['STD-YM']]
df_first
```

```python
df_11 = pd.DataFrame()
```

```python
for i in range(0, len(df_first), 1) : 
    if df_first['STD-YM'][i].endswith('11') :
        df_11 = pd.concat([df_11, df_first.iloc[[i]]])
df_11
```

```python
df_final = pd.concat([df_4,df_11]).sort_index()
df_final.head()
```

```python
df_final['return'] = 0
```

```python
for i in range(0, len(df_final), 1) :
    if df_final['STD-YM'][i].endswith('04') :
        df_final['return'][i] = (df_final['Close'][i] - df_final.shift(1)['Open'][i]) / df_final.shift(1)['Open'][i]

df_final
```

```python
df_rtn = df_final.loc[df_final['return'] !=0][['return']]
df_rtn
```

```python
df_rtn['total_return'] = df_rtn['return'] + 1
df_rtn
```

```python
df_rtn['total_return'].cumprod()
```

```python
from datetime import datetime
from dateutil.relativedelta import relativedelta
```

```python
start = datetime(year = 2000, month= 11, day = 1)
start
```

```python
end = start + relativedelta(months = 5) # 반년뒤 5월초가 아니라 4월 말 종가가 필요하기 때문에 5개월만 뒤로 가자 
end
```

```python
print(start.strftime('%Y-%m'), end.strftime('%Y-%m'))
```

```python
def six_month(df, start = 2000, end = 2019, month = 11) : 
    if 'Date' in df.columns :
        df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
        df.set_index('Date', inplace=True)
        
    # 누적수익률 변수를 생성
    acc_rtn = 1
    
    for i in range(start, end):
         
        start_ = datetime(year = i, month=month, day = 1)
        end_ = start_ + relativedelta(months = 5)

        buy_mon = start_.strftime('%Y-%m')
        sell_mon = end_.strftime('%Y-%m')

        buy = df.loc[buy_mon].iloc[0]['Open']
        sell = df.loc[sell_mon].iloc[-1]['Close']

        rtn = (sell - buy)/buy + 1

        acc_rtn *= rtn

    return acc_rtn
```

```python
df = pd.read_csv('../csv/AMZN.csv')
```

```python
six_month(df)
```

```python


for i in range(1, 13, 1) :
    print(f"{i} 월 : ", six_month(df, month=i))
```

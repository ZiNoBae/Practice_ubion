```python
kospi = pd.read_excel('../data/kospi.xlsx', usecols= [0, 1, 4, 5, 6, 7], parse_dates = [0]) # usecols -> 몇번째 컬럼들을 쓸건지 # parse_dates -> 몇번째를 시계열로 바꿀건지
```

```python
kospi.info()
```

```python
how = {
    '시가' : 'first',
    '고가' : 'max',
    '저가' : 'min',
    '종가' : 'last',
    '거래량' : 'sum'
}

group_data = kospi.groupby(pd.Grouper(key='일자', freq = 'W')).agg(how)
    # Grouper - 나눌때 기준 (분기별, 월별, 일별) -> 보통 groupby(pd.Grouper) 같이 쓰임 
```

```python
# weekday
    # 0 = 월 부터 시작

kospi.loc[0, '일자'].weekday()
```

```python
# 주 week라는 컬럼 만들어보자 

for i in kospi.index : 
    kospi.loc[i,'요일'] = kospi.loc[i, '일자'].weekday()
```

```python
kospi
```

```python
kospi = kospi.sort_values('일자').set_index('일자')
```

```python
# 한주 시작 의 조건?
    # 전의 요일 데이터보다 현재의 요일 데이터가 작거나 같은 경우 

kospi.loc[kospi['요일'] <= kospi.shift(1)['요일']].value_counts('요일')
```

```python
# 한주 끝의 조건 ?
    # 다음날의 요일이 현재 요일보다 작거나 같은 경우 
kospi.loc[kospi['요일'] >= kospi.shift(-1)['요일']].value_counts('요일')
```

```python
# 주 시작날의 시가 & 주 종료날의 종가로 주별 수익률 구해보자 
week_start = kospi.loc[kospi['요일'] <= kospi.shift(1)['요일']]
```

```python
week_start['trade'] = 'buy'
```

```python
week_start
```

```python
week_end = kospi.loc[kospi['요일'] >= kospi.shift(-1)['요일']]
```

```python
week_end['trade'] = 'sell'
```

```python
df_week = pd.concat([week_start, week_end]).sort_values('일자')
```

```python
df_week
```

```python
df_week['return'] = 1
```

```python
df_week
```

```python
for i in df_week.index :
    if df_week.loc[i,'trade'] == 'sell' : 
        df_week.loc[i, 'return'] = (df_week.loc[i,'종가'] - df_week.shift(1).loc[i,'시가'])/ df_week.shift(1).loc[i, '시가'] + 1

df_week
```

```python
df_zn = df_week.loc[df_week['return'] != 1 ][['return']] 
```

```python
# Moon 

kospi['rtn'] = 1
buy = kospi.iloc[0,1] 
#print(buy)

for i in kospi.index : 
    # 1월 3일 주의 마지막이기 때문에 buy가 존재할수가 없어 미리 첫날의 시기를 지정
 
    # 주의 시작날의 시가를 buy에 지정
    if kospi.shift(1).loc[i, '요일'] >= kospi.loc[i, '요일'] : 
        buy = kospi.loc[i, '시가']
    # 주의 마지막날 종가를 sell에 지정
    elif kospi.shift(-1).loc[i,'요일'] <= kospi.loc[i, '요일'] : 
        sell = kospi.loc[i, '종가']
        rtn = (sell - buy) / buy + 1
        kospi.loc[i, 'rtn'] = rtn
```

```python
# group_data -> 7일별 자르는 방식 의 수익률 과 우리가 한것과 비교해보자

group_data = kospi.groupby(pd.Grouper(key = '일자', freq = 'W')).agg(how)
```

```python
group_data
```

```python

group_data['수익률'] = (group_data['종가'] - group_data['시가']) / group_data['시가'] + 1
```

```python
for i in group_data.index : 
    group_data['요일'] = i.weekday()

group_data
```

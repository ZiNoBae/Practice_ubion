# 문제
1. emp table에서 SAL 1000 이상이고 3000 미만인 사원정보를 출력
2. emp, dept를 조인결합 left 결합한 후 부서의 위치가 'NEW YORK'인 데이터 출력

```python
import pandas as pd
import numpy as np
import mod_sql
```

```python
db = mod_sql.Database( 'localhost', 'root', '1234', 'ubion7', 3306)
```

```python
sql = """
    select
    *
    from
    emp
    where
    `SAL` >= 1000
    and
    `SAL` < 3000
"""

db.sql_query(sql)
```

```python
sql = """
    select
    *
    from
    emp
    left join
    dept
    on
    emp.DEPTNO = dept.DEPTNO
    where
    `LOC` = 'NEW YORK'
"""

db.sql_query(sql)
```

## 문제 Cont'd

3. emp, dept를 right join 결합
4. 부서번호가 40인 데이터만 출력

```python
sql = """
    select
    *
    from
    emp
    right join
    dept
    on
    emp.DEPTNO = dept.DEPTNO
    where
    dept.DEPTNO = 40
"""

# where 이하 조건은 left결합이면 left류, right결합이면right류 권장 
db.sql_query(sql)
```

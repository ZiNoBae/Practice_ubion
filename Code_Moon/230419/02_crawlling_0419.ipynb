```python
import requests
from bs4 import BeautifulSoup as bs

import imp
imp.reload(requests)
```

```python
# 실제 사이트에서 html정보를 받아와서 변수에 저장
data = requests.get('https://finance.naver.com/marketindex/exchangeList.naver').text
```

```python
# BeautifulSoup을 이용해서 데이터를 파싱

soup = bs(data, 'html.parser')
```

```python
print(soup)
```

```python
# find는 html 안에서 해당 태그의 첫번째 정보를 출력
soup.find('tbody')
```

```python
# find_all() 해당 태그의 전체 정보를 리스트의 형태로 출력

data_list = soup.find_all('td')
```

```python
# data_list를 이용하여 2차원 배열로 새로운 리스트를 생성
# 1차원 리스트 생성
# [[7개씩],[7개씩],[7개씩],[] ...] -> 2차원 리스트

_list = []

for i in data_list :
    x = i.get_text()
    x = x.replace('\n', '')
    x = x.strip('\t')
    _list.append(x)
_list    

# 까지가 1차원 리스트 
```

```python
type(len(_list)//7)  # 몫만 추출하기때문에 dtype : int
```

```python
list(range(0, (len(_list)//7),1))
```

```python
result = []
for i in range(0, len(_list)//7, 1) :
    x= []
    for j in range(i*7, (i+1)*7, 1) :
        x.append(_list[j])
    result.append(x)
result
```

```python
# 조금 더 simple하게 뽑아보자 
# list[x] : 리스트의 x위치
# list[x:y]: 리스트의 x 위치부터 y 전까지 

_list[]
```

```python
result = []
for i in range(len(_list)//7) : 
    result.append(_list[i*7 : (i+1)*7])
result
```

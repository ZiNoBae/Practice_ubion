- 행동 
- 반복적인 코드를 실행하는 경우, 코드를 반복해서 사용시 
    - 코드 라인 증가
    - 가독성 하락
- 코드가 길어져 error 발생확률이 증가
- def 키워드를 이용하여 함수를 선언



```python
# 함수를 생성
# 인자의 값이 숫자의 형태가 아니면 '숫자의 형태가 아니다' 라고 출력
# 숫자인 경우 짝수이면 '짝수'
# 홀수이면 '홀수' 로 출력하는 함수

def func(x) :
    if type(x) == int :
        if x%2 == 0 :
            return '짝수'
        else :
            return '홀수'
    elif type(x) == float:
        return '실수의 형태'

    else :
        return '숫자의 형태가 아니다'
    
```

```python
a = 128937
b = 1020
c = 0.123
print(func('가나다'))
print(func(a))
print(func(b))
print(func(c))
```

```python
# 함수 생성 
# 매개변수 2개
# 두 매개변수 사이 값들의 합 리턴
# 3,5 -> 3+4+5 -> 12 리턴 할 수 있게
# 두 매개변수의 대소관계는 상관없이 사이의 값들이 합

# ZiNo
def func_2(x,y) :
    return (x+y)*(abs(y-x)+1)/2
```

```python
# Moon

# I. for문
def func2(x,y) :
    result = 0
    if x < y :
        for i in range(x,y+1,1) :
            result += i
    else :
        for i in range(y, x+1, 1) :
            result += i

    return result

    
```

```python
func2(1,10)
```

```python
func2(10,1)
```

```python
# Moon II. max

def func3(x,y) :
    result = 0

    for i in range(min(x,y), max(x,y)+1, 1) :
        result += i

    return result
```

```python
func3(10,1)
```

```python
# 함수 생성
# 매개변수는 1개
# 인자값이 12345 로 들어가면, 
# 1+2+3+4+5 출력이 되도록 함수를 구성

def func_4(x) :
    
    result = 0

    for i in range(0, len(str(x)), 1) :
        result += int(str(x)[i])

    return result
```

1. 상품 A의 item_price의 값은 결측치 중 상품이 상품A면 이 값을 치환
2. 나머지 상품들도 같은 작업을 반복

```python
# 상품A의 item_price 출력
# 행의 조건 ? 결측치가 아니어야함 & item_name이 상품A이다
# 열의 조건 ? item_price

# 조건1. 결측치가 아니다
flg_null = uriage['item_price'].isna()
# uriage.isna().any(1)   <- 얘가 위보다 느려 
```

```python
# 결측치인 값들 출력

uriage.loc[flg_null]
```

```python
# item_name이 상품A인 경우만 출력

uriage.loc[uriage['item_name']=='상품A']
```

```python
# 상품A 이고 결측치 item_price가 결측치인 경우 (위 2개셀 조건 both적용)
# 조건 ()에 넣으램

uriage.loc[(flg_null) & (uriage['item_name']=='상품A')]
```

```python
# 상품A 이고 item_price가 결측치가 아닌 경우

price = uriage.loc[(uriage['item_name']=='상품A') &\
            ~(flg_null), 'item_price'].value_counts().index[0]
```

```python
# 결측치 값을 치환하는 함수
# fillna(n) : 결측치인 곳을 채워넣는다 라는 의미

uriage.loc[(flg_null) & (uriage['item_name']=='상품A')].fillna(price)
```

```python
uriage.loc[(flg_null) &\
           (uriage['item_name']=='상품A'), 'item_price'] = price
```

```python
# item_name 리스트 출력 ->  반복문에 얘만 쓰려고

name_list = uriage['item_name'].unique()
name_list
```

```python
for i in name_list : 
    #item_price의 값 구하기
    # item_name이 i와 같고, 결측치가 아닌 데이터에서 price를 출력
    price = uriage.loc[~(flg_null)&\
                       (uriage['item_name']== i), 'item_price']\
                       .value_counts().index[0]  #여기까지 나온 값을 결측치에 넣어줘야지
    # item_name이 i와 같고 결측치인 데이터에서 item_price라는 변수를 삽입 
    uriage.loc[(flg_null)&\
               (uriage['item_name']== i ), 'item_price'] = price
```

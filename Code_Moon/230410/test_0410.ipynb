# beautifulsoup 라이브러리

1. html tag의 데이터를 쉽게 추출하기 위한 라이브러리 
2. 웹의 구조를 어느정도 인지를 한 상태에서 라이브러리 사용하면 쉽게 데이터 추출 
3. 파서(parser) 활용하여, 파이썬에서 접근이 쉽게 객체 형태로 제공

```python
from bs4 import BeautifulSoup as bs
```

```python
# html 형태로 파싱

soup = bs(data, 'html.parser')
```

```python
print(soup)
```

- 태그의 이름을 사용해서 검색 
    - soup[태그명] : 해당 태그의 첫번째 정보를 출력
    - soup[태그명][속성] : 해당 태그의 첫번째 속성에 대한 값을 출력

```python
print(soup.title)
print(soup.title.name)
print(soup.title.string)
```

```python
print(soup.img) # img -> 이미지 태그 
print(soup.img['alt'])
```

- find()     <- beautifulsoup 안에 있는 find 
    - HTML 문서 안에서 해당 태그의 첫번째 정보 출력
    - find(속성 = 값) : HTML 문서 안에서 해당 속성과 일치하는 값을 가진 태그의 첫번째 정보출력
    

```python
print(soup.find('a'))
```

```python
print(soup.find(id = 'newsstand'))
```

- find_all()
    - HTML 문서 안에서 해당 태그에 대한 정보를 모두 출력(데이터 형태 : list)
    - limit 매개변수 : 리스트의 길이 지정

```python
url = "https://finance.naver.com/"

html_data = requests.get(url).text
```

```python
soup = bs(html_data, 'html.parser')   # 데이터 파싱 작업까지가 기본작업
```

```python
data = soup.find('table', attrs={
    'class' : 'tbl_home'
})

```

```python
data2 = data.find("tbody")
```

```python
_list = data2.find_all('th')
```

```python
result = []

for i in _list : 
    result.append(i.get_text())

pp.pprint(result)
```

```python
# 주가 종목 오른애들 class = up 이네 얘네 종목만 찾아볼까

data3 = data2.find_all('tr', attrs={
    'class' : 'up'
})
```

```python
# td 리스트 

_list3 = []

for i in data3 : 
    _sample_list = []
    _sample_list.append(i.find('th').get_text())
    _list4 = i.find_all('td')
    for j in _list4 :
        _sample_list.append(j.get_text())
    _list3.append(_sample_list)

_list3
```

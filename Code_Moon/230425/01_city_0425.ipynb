```python
# Moon

# 공백지우기 : 양쪽 공백 지우기
df['분양가격'] = df['분양가격'].str.strip()

# , comma 지우기
df['분양가격'] = df['분양가격'].str.replace(",", "")

# - bar 지우기
df['분양가격'] = df['분양가격'].str.replace('-', '')

# 공백 value -> 0 으로 대체
df.loc[df['분양가격'] == "", '분양가격'] = 0

# 결측치 0으로 대체 
df['분양가격'].fillna(0, inplace=True)
```

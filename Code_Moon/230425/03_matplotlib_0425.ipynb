## subplot()
1. 한 영역에 그래프를 여러 개 그려주기 위한 함수 
2. subplot(행의 개수, 열의 개수, 위치값)

```python
plt.subplot(1, 2, 1)
plt.plot([1,2,3,4])

plt.subplot(1,2,2)
plt.plot([4,3,2,1])
```

```python
fig, axes = plt.subplots(2,2) # fig -> 사이즈 지정할때 옵션 지정값 넣어준대


axes[0][0].plot([1,2,3,4])
axes[0][1].plot([4,3,2,1])
axes[1][0].plot([1,2,2,1])
axes[1][1].plot([3,4,2,1])

plt.show()
```

```python
# 주의 : x축 값의 갯수와 y축 값의 개수 일치시켜줄 것 ! -> 불일치면, Error

x= [1,2,3]
y = [2,4,6,8]

plt.plot(x,y)
plt.show()
```

```python
# x 축의 이름과 y축의 이름, 그래프의 이름지정

plt.plot([1,2,3,4])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('title')
plt.show()
```

```python
# 범례 추가 

plt.plot([1,2,3,4], label = 'acending')
plt.plot([4,3,2,1], label = 'descending')
plt.legend(loc = 'upper right') # 숫자도 가능, [숫자,숫자], 'right' 등등 가능
plt.show()
```

```python
# 히스토그램 -> 데이터의 개수를 표현하는 바 그래프

data = [1, 2,2,2, 3,3, 4,4,4,4]

plt.hist(data, bins= 4) # bins = 막대 갯수, 범위를 나눠준 후 해당 범위에 총 갯수 
plt.show()
```

```python
mean_kospi_2020 = kospi.loc['2020'].resample('MS').mean()
mean_kospi_2020
```

```python
# x축은 'XXXX년XX월' 표시 
mean_kospi_2020.index[0].strftime('%Y년%m월')
```

```python
# 반복문으로 년월 리스트 뽑기
x = []

for  i in mean_kospi_2020.index : 
    x.append(i.strftime('%Y년%m월'))

x
```

```python
# map함수로 년월 리스트 뽑기

a= []
list(map(lambda x : a.append(x.strftime('%Y년%m월')), mean_kospi_2020.index))

a
```

# 문제

1. test_csv 폴더안에 있는 오프라인.csv 로드
2. 컬럼의 이름을 ['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']으로 변경
3. 기준년월을 시계열 데이터로 변경
4. index를 기준년월로 변경 
5. 온라인.csv파일도 위의 작업을 반복
6. 온라인에 업종 컬럼을 추가하여 '온라인'으로 값을 지정
7. 두개의 데이터 프레임을 유니언 결합
8. 기준년월, 업종별 총 이용금액의 합계 -> 선 그래프로 시각화

```python
import pandas as pd
```

```python
# 1. 로드 

offline = pd.read_csv('../test_csv/오프라인.csv', encoding='euc_kr')
origin_ = offline.copy()
```

```python
offline.head()
```

```python
# 2. 컬럼 이름 변경

offline.columns = ['기준년월', '업종', '성별', '연령대', '총이용건수', '총이용금액']
offline.head()
```

```python
# 3. 기준년월 -> 시계열 

offline['기준년월'] = pd.to_datetime(offline['기준년월'], format = '%Y%m')
offline.info()
```

```python
# 4. index '기준년월'

offline.set_index('기준년월', inplace= True)
offline.head()
```

```python
# 1. 온라인 로드

online = pd.read_csv('../test_csv/온라인.csv', encoding = 'euc_kr')
_origin = online.copy()
```

```python
# 2. 컬럼 이름 변경

online.columns = ['기준년월', '성별', '연령대', '총이용건수', '총이용금액']
online.head()
```

```python
# 3. 시계열

online['기준년월'] = pd.to_datetime(online['기준년월'], format = '%Y%m')
online.info()
```

```python
#4. index 기준년월

online.set_index('기준년월', inplace=True)
online.head()
```

```python
offline
```

```python
online
```

```python
# 5. 온라인에 업종 컬럼 추가

online['업종'] = '온라인'
```

```python
online.head()
```

```python
_df = pd.concat([offline, online], axis = 0)  # 매개변수 ignore_index = True 하면, 우리가 만든 기준년월 인덱스 사라짐 주의
_df
```

```python
_df.sort_index(inplace = True) # 안해도 상관없음
```

```python
df = _df.groupby(['기준년월', '업종']).sum()[['총이용금액']]
df
```

```python
import platform

if platform.system() == 'Dawin' :
    plt.rc('font', family = 'AppleGothic')
else : 
    plt.rc('font', family = 'Malgun Gothic')
```

```python
df_pivot = pd.pivot_table(df, index = '기준년월', columns = '업종', values='총이용금액')
```

# Buy and Hold 함수로 생성

1. 매개변수는 1개(dataframe)
2. 결측치, 무한대, 음의 무한대 제거 
3. 수정종가만 있는 데이터프레임으로 변경 
4. 일별 수익을 파생변수 생성
5. 누적수익률 파생변수 생성
6. 데이터프레임을 리턴

```python
def ZoneNaHold(x) : 
    x = x.loc[~x.isin([np.nan, np.inf, -np.inf]).any(1)]
    x_price = x[['Adj Close']]
    x_price['daily_rtn'] = x_price['Adj Close'].pct_change()
    x_price['cumulative_rtn'] = (1 + x_price['daily_rtn']).cumprod()



    return x_price
```

```python
# 함수에 기준이 되는 컬럼을 매개변수 추가생성 -> 국내 data는 Adj Close가 아닐수도 ?! Close 일수도?
# 매개변수 2개 구매 시기와 판매시기 : start, end 추가 

def bnh(x, col, start, end) : 
    x = x.loc[~x.isin([np.nan, np.inf, -np.inf]).any(1)]
    x = x[[col]]
    x['daily_rtn'] = x[col].pct_change()
    x['cumulative_rtn'] = (1 + x['daily_rtn']).cumprod()



    return x
```
